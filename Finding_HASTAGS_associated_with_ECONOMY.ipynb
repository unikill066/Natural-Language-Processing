```python
# Imports
import re, nltk, numpy as np, pandas as pd

# Loading the raw tweets data
df = pd.read_csv("raw_tweets_economy.txt", sep="\t")
df.columns = ['Line No.', 'Tweets'] # changing column names to meaningful entries
df.head()
```

```python
# Testing
sentence = "RT @Emile1chedid: #Lebanon #Beirut @politics #.."
hastags = re.findall(r"#[a-zA-z0-9]+", sentence)
print(hastags)
```

```python
hash_list = [] # defining an empty list

def hastag_list(series):
    """
    This function takes a series, array, list, etc. and convert the data into list of lists which ever entry has a HASHTAG.
    returns: list of lists contentained of hastags along with the tag. 
    """
    for i in series:
        temp = re.findall(r"#[a-zA-z0-9]+", i)
        hash_list.append(temp)
    return hash_list
    
hash_list = hastag_list(df['Tweets']) # re-defining the same variable 
hash_list[0:3] 
```

```python
# Findind out only the hashtags which are assiciated with #ECONOMY in the list of lists

eco_hash_list = [] # defining an empty list

for item in hash_list:
    for j in item:
        if "economy" in j:
            eco_hash_list.append(item)
            continue

eco_hash_list
```

```python
# All the wordings used in the tweets

words = []
for i in df['Tweets']:
    temp = re.findall(r"#[a-zA-z0-9]+", i)
    for j in temp:
        if j[1:] not in words: words.append(j[1:])
        
words
```
